```python
import pandas as pd
import matplotlib.pyplot as plt  
```

```python
turnstiles_daily = pd.read_csv('turnstiles_daily_clean.csv').drop('Unnamed: 0', axis = 1)
```

```python
turnstiles_daily['DATE'] = pd.to_datetime(turnstiles_daily.DATE)
```

# MTA Station Traffic

### MVP overview

- **Goal:** An environmental nonprofit is looking to collect signatures in support of legislature that will provide additional funding for light rail improvements. Knowing which stations are busiest and when will help their small team of canvassers plan when to be where and help ensure that they encounter as many people as possible. As such, this analysis seeks to  identify the MTA stations with the highest traffic and then determine on what days those stations are anticipated to be busiest. 


- **Process:** Data exploration to reach this conclusion entailed aggregation of daily commuter data at the turnstile level and then to the MTA station level.


- **Preliminary visualization:** The below visualizations identify which days of the week and stations have the most traffic, on average. These averages were calculated using data from all of 2021. 


- **Preliminary conclusions:** Weedays look to be busier than weekend days. JUNCTION BLVD appears to be the busiest station on average. 

```python
bar_data = (turnstiles_daily
 .groupby("DOW")
 .DAILY_ENTRIES.mean()
 .reset_index())

plt.bar(bar_data.DOW, bar_data.DAILY_ENTRIES)
plt.title("Average entries per station")
plt.ylabel('Number of People')
plt.xticks([0,1,2,3,4,5,6], ['Mon', 'Tues', 'Wed','Thurs','Fri','Sat','Sun']);
```

```python
other_bar = (turnstiles_daily.groupby(["STATION"])
 .DAILY_ENTRIES.mean()
 .reset_index()
 .sort_values('DAILY_ENTRIES', ascending = False)
 .head(10))

plt.bar(other_bar.STATION, other_bar.DAILY_ENTRIES)
plt.xticks(rotation = 90)
plt.title('10 Stations with Highest Traffic on Average')
plt.ylabel('Number of entries per day');
```
